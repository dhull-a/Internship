#### Q.  Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
imdb =requests.get('https://www.imdb.com/chart/moviemeter/')
```

```python
imdb
```

```python
#display the page source code
imdb.content
```

```python
soup2=BeautifulSoup(imdb.content, "html.parser")
```

```python
print(soup2.prettify())
```

```python
#scrappping movie names

movie_name=soup2.find_all('td',class_='titleColumn')
movie_name
```

```python
#parse movie name

movies = []
for movie in movie_name:
    movie=movie.get_text().replace('\n',"")
    movie= movie.strip(" ")
    movies.append(movie)
movies
```

```python
#scrappping ratings

movie_ratings=soup2.find_all('td',class_='ratingColumn imdbRating')
movie_ratings
```

```python
#parse ratings

ratings = []
for rating in movie_ratings:
    rating=rating.get_text().replace('\n',"")
    rating= rating.strip("\n")
    ratings.append(rating)
ratings
```

```python
#scrappping year

movie_year=soup2.find_all('span',class_='secondaryInfo')
movie_year
```

```python
#parse ratings

years = []
for year in movie_year:
    year=year.get_text().replace('\n',"")
    years.append(year)
years
```

```python
print(len(movie_name), len(movie_ratings), len(movie_year))
```

```python
#Making the datframe

data=pd.DataFrame()
data['Movie Names'] =movie_name
data['Ratings'] =movie_ratings
data.head()
```
